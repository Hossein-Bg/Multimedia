# Read an Image

# Convert the image to gray

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Create the Salt & Pepper noise for the Gray Image

**If you want to apply the specific amount of noise, Please comment the line 3 and uncomment the line 4

```python
def salt_pepper(image):
  w, h = image.shape
  random_pix = np.random.randint(0, 10000)
  # random_pix = 10000

  # Salt noise
  for i in range(random_pix):
    ycord = np.random.randint(0, w-1)
    xcord = np.random.randint(0, h-1)

    image[ycord, xcord] = 255

  # pepper noise
  for i in range(random_pix):
    ycord = np.random.randint(0, w-1)
    xcord = np.random.randint(0, h-1)

    image[ycord, xcord] = 0

  return image
```

# Save the output

```python
img_item = '/content/drive/MyDrive/multimedia-14023-Hossein Baghdadi-980201110022/2/Output/Salt_Pepper.jpg'
cv2.imwrite(img_item, salt_pepper(gray))
```
