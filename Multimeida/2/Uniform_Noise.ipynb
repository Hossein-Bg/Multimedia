# Tools

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
```

# Read an Image

```python
img = cv2.imread('/content/drive/MyDrive/multimedia-14023-Hossein Baghdadi-980201110022/2/wp9077851.jpg')
cv2_imshow(img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Resize the Image

```python
resized = cv2.resize(img, (600, 600))
cv2_imshow(resized)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Convert the image to gray

```python
gray = cv2.cvtColor(resized, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Create the Uniform noise for the Gray Image

```python
uniform_noise = np.random.uniform(0, 500, gray.shape) # low=0 & high=500
```

# Showing the noise

```python
cv2_imshow(uniform_noise)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Add the noise to the gray Image

```python
uniform_noise_img = gray + uniform_noise  
```

# Save the output

```python
img_item = '/content/drive/MyDrive/multimedia-14023-Hossein Baghdadi-980201110022/2/Output/Uniform_noise.jpg'
cv2.imwrite(img_item, uniform_noise_img)
```

# Showing the Resized image, Gray image and the output

```python
cv2_imshow(gray)
cv2_imshow(uniform_noise_img.astype(np.uint8))
          
cv2.waitKey(0)
cv2.destroyAllWindows()
```
