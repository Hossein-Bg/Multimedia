# Tools

```python
import cv2 
import numpy as np
from google.colab.patches import cv2_imshow
```

# Read an Image

```python
img = cv2.imread('/content/drive/MyDrive/multimedia-14023-Hossein Baghdadi-980201110022/2/124306-Iran-Tehran-city-Azadi_Square-monochrome-architecture.jpg')
cv2_imshow(img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Resize the Image


```python
resized = cv2.resize(img, (600, 600))
cv2_imshow(resized)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Convert the image to gray

```python
gray = cv2.cvtColor(resized, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Create the Gaussian noise for the Gray Image

```python
mean = 0
var = 500
sigma = np.sqrt(var)

Gaussian_noise = np.random.normal(loc=mean, scale=sigma, size=gray.shape)
```

# Showing the noise

```python
cv2_imshow(Gaussian_noise)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Add the noise to the gray Image

```python
Gaussian_noise_img = gray + Gaussian_noise
```

# Save the output

```python
img_item = '/content/drive/MyDrive/multimedia-14023-Hossein Baghdadi-980201110022/2/Output/Gaussian_noise.jpg'
cv2.imwrite(img_item, Gaussian_noise_img)
```

# Showing the Gray image and the output

```python
cv2_imshow(gray)
cv2_imshow(Gaussian_noise_img.astype(np.uint8))

cv2.waitKey(0)
cv2.destroyAllWindows()
```
