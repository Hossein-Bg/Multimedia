## BackGround & Foreground

# Tools

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import os
import random
```

# Read the Background Image

```python
backg = cv2.imread('/content/drive/MyDrive/multimedia-14023-Hossein Baghdadi-980201110022/1/background/13DEVELOPMENT-slide-I3QM-superJumbo.jpg')
cv2_imshow(backg)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
foreground_dir = "/content/drive/MyDrive/multimedia-14023-Hossein Baghdadi-980201110022/1/foreground"

for foreground_file in os.listdir(foreground_dir):
    # Read the foreground image
    foreg = cv2.imread(os.path.join(foreground_dir, foreground_file))
    # cv2_imshow(foreg)

    # Resize the foreground images to match with size of background image
    foreg_resized = cv2.resize(foreg, (backg.shape[1], backg.shape[0]))
    # cv2_imshow(foreg_resized)

    # find the random cordination for foreground images
    x = random.randint(0, backg.shape[1] - foreg.shape[1])
    y = random.randint(0, backg.shape[0] - foreg.shape[0])

    # make a mask for foreground image
    gray = cv2.cvtColor(foreg, cv2.COLOR_BGR2GRAY)
    _, mask = cv2.threshold(gray, 10, 255, cv2.THRESH_BINARY)

    # make a mask for background image
    inv_mask = cv2.bitwise_not(mask)

    # Extract the foreground region
    foreg = cv2.bitwise_and(foreg, foreg, mask=mask)

    # Extract the background region
    backg_roi = backg[y:y+foreg.shape[0], x:x+foreg.shape[1]]
    backg_roi = cv2.bitwise_and(backg_roi, backg_roi, mask=inv_mask)

    dst = cv2.add(backg_roi, foreg)

    backg[y:y+foreg.shape[0], x:x+foreg.shape[1]] = dst

# Final output
cv2_imshow(backg)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
