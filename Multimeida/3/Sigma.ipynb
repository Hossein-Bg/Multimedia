```python
import numpy as np
import matplotlib.pyplot as plt

def add_gaussian_noise(image, sigma):
    noise = np.random.normal(0, sigma, image.shape)
    noisy_image = image + noise
    return noisy_image

# Load and display an example image
image = plt.imread('/content/drive/MyDrive/multimedia-14023-Hossein Baghdadi-980201110022/3/5502233.jpg')
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')
plt.show()

# Generate and display noisy images with different sigma values
sigma_values = [10, 25, 50, 75, 100]

for sigma in sigma_values:
    noisy_image = add_gaussian_noise(image, sigma)
    plt.imshow(noisy_image)
    plt.title('Noisy Image (sigma = {})'.format(sigma))
    plt.axis('off')
    plt.show()
```
