# Tools

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

# Read an Image

```python
img = cv2.imread('/content/drive/MyDrive/multimedia-14023-Hossein Baghdadi-980201110022/4/malcomx.jpeg')
cv2_imshow(img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Convert the image to gray

```python
gray = np.mean(img, axis=2)
gray
```

# Plot Image as Signal

```python
plt.plot(gray.flatten(), color='green', linewidth=2)
plt.title('Image as Signal')
plt.show()
```

# Specify a kernel size for our Average Filter

```python
kernelSize = 7
kernel = np.ones((kernelSize, kernelSize)) / (kernelSize ** 2)
kernel
```

# Create and apply Average Filter on gray Image

```python
average_filter = np.zeros_like(gray)
for i in range(kernelSize // 2, (gray.shape[0] - kernelSize) // 2):
  for j in range(kernelSize // 2, gray.shape[1] - kernelSize // 2):
    average_filter[i, j] = np.sum(gray[i - kernelSize // 2:i + kernelSize // 2 + 1, j - kernelSize // 2:j + kernelSize // 2 + 1] * kernel)
```

# Plot the Average Filtered Image as Signal

```python
plt.plot(average_filter.flatten())
plt.title('Average Filter on Image as Signal')
plt.show()
```

# Showing the Image and the Average Filtered Image

```python
plt.plot(gray.flatten(), color='green', label='Image')
plt.plot(average_filter.flatten(), label='Average Filter')
plt.legend(loc='best')
plt.title('Image and Average Filter Signals')
plt.show()
```
